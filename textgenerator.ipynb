```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# Define your text data
text = "This is a sample text for text generation using LSTM."

# Preprocess the text data
chars = sorted(list(set(text)))

ci = {char: i for i, char in enumerate(chars)}
ic = {i: char for i, char in enumerate(chars)}
n = len(chars)
seqlength = 10


dataX = []
datay = []

for i in range(0, len(text) - seqlength):
    seq_in = text[i:i+seqlength]
    seq_out = text[i+seqlength]
    dataX.append([ci[char] for char in seq_in])
    datay.append(ci[seq_out])

#reshape X so that it can be feeded into  the LSTM model 

X = np.reshape(dataX, (len(dataX), seqlength, 1))
X = X / float(n) #normalize the data 
y = np.eye(n)[datay] #creating a matrix and values in it alone will be one 

# Define and train the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(n, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=100, batch_size=16)

# Generate text using the trained model
start_seq = "This is a "
generated_text = start_seq

for _ in range(50):
    x = np.reshape([ci[char] for char in start_seq], (1, len(start_seq), 1))
    x = x / float(n)
    prediction = model.predict(x, verbose=0)
    index = np.argmax(prediction)
    result = ic[index]
    generated_text += result
    start_seq = start_seq[1:] + result

print(generated_text)
```
